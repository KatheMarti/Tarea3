```python
#2. Preprecesar los datos limpiándolos, tratando valores faltantes y transformándolos según sea necesario.
    
```

```python
#Revisión de datos o valores faltantes, dentro de columna datos cuenteme si hay valores en cero
print ('Cantidad de Registros con cero en Nivel de trestbps :', df['trestbps'].isin([0]).sum())
print ('Cantidad de Registros con cero en Nivel de fbs :', df['fbs'].isin([0]).sum())
print ('Cantidad de Registros con cero en Nivel de restecg :', df['restecg'].isin([0]).sum())
print ('Cantidad de Registros con cero en Nivel de thalach :', df['thalach'].isin([0]).sum())
print ('Cantidad de Registros con cero en Nivel de oldpeak :', df['oldpeak'].isin([0]).sum())
print ('Cantidad de Registros con cero en Nivel de slope :', df['slope'].isin([0]).sum())
```

```python
#Reemplazar todos los ceros con el valor medio (promedio) de la columna
df['trestbps']=df['trestbps'].replace(0,df['trestbps'].mean())
df['fbs']=df['fbs'].replace(0,df['fbs'].mean())
df['restecg']=df['restecg'].replace(0,df['restecg'].mean())
df['thalach']=df['thalach'].replace(0,df['thalach'].mean())
df['oldpeak']=df['oldpeak'].replace(0,df['oldpeak'].mean())
df['slope']=df['slope'].replace(0,df['slope'].mean())
```

```python
#Revisión nuevamente de datos faltantes
print ('Cantidad de Registros con cero en Nivel de trestbps :', df['trestbps'].isin([0]).sum())
print ('Cantidad de Registros con cero en Nivel de fbs :', df['fbs'].isin([0]).sum())
print ('Cantidad de Registros con cero en Nivel de restecg :', df['restecg'].isin([0]).sum())
print ('Cantidad de Registros con cero en Nivel de thalach :', df['thalach'].isin([0]).sum())
print ('Cantidad de Registros con cero en Nivel de oldpeak :', df['oldpeak'].isin([0]).sum())
print ('Cantidad de Registros con cero en Nivel de slope :', df['slope'].isin([0]).sum())
```

```python
#Identificación de datos atípicos
plt.figure(figsize=(6,3))
sns.boxplot(x=df['chol'])
plt.title('Identificación de Datos Faltantes y Atípico', fontsize=10 )
```

```python
#Conteo de datos atípicos
nivel_minimo=20
nivel_maximo=240
Total_Atipicos=((df['chol']<nivel_minimo)|(df['chol']>nivel_maximo)).sum()
print ('El total de datos atípicos es {}'.format(Total_Atipicos))
```

```python
#filtrar datos atípicos
df=df[(df['chol']<nivel_maximo)&(df['chol']>nivel_minimo)]
```

```python
#validar si se removieron datos atípicos
Total_Atipicos=((df['chol']<nivel_minimo)|(df['chol']>nivel_maximo)).sum()
print ('El total de datos atípicos es {}'.format(Total_Atipicos))
```

```python
#3. Seleccionar las características más relevantes para entrenar el modelo utilizando selección de características
#total de personas con la condición de enfermedad cardiaca
df['condition'].value_counts()
```

```python
#4. Dividir dataset en Train y Test para evaluar correctamente el modelo
#procentaje de datos para entrenamiento (train)=75% y para prueba (test) =25%
#se crean variables X y Y para almacenar la información del entrenamiento

X= df.drop('condition', axis=1)# se cargan los datos de traina sin los datos de columna condition
Y=df['condition']# se cargan los datos de train con los datos de la columna codition

# la función train_test_split crea una división de un conjunto de datos en dos bloques uno de entrenamiento y otro de prueba(train and test)
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.25, stratify=Y, random_state=0)
```

```python
print (X_train)
```

```python
print (X_test)
```

```python
#5.Entrenar el modelo configurado los diferentes hiperparámetros
#Constriut el modelo de regresión logístico
modelo_regresion=LogisticRegression(solver='liblinear')# liblinear es una libreria para resolver problemas lineales de clasificación y regresión 
modelo_regresion.fit(X_train , Y_train)

#Crea modelo de prediccion
Prediccion=modelo_regresion.predict(X_test) # La función predict predice los valoes a partir del conjunto de datos de validación
```

```python
X_test.head(10)
```

```python
print (Prediccion)
```

```python
#6. Evaluar el desempeño del modelo en el conjunto de Test con métricas como precisión, recall, F1-score

#Desempeño del modelo de regresión
print(classification_report(Y_test, Prediccion))
```

```python
#7. Realizar las diferentes gráficas que permitan visualizar los resultados del modelo

#Matrix de confusión del modelo

cm=confusion_matrix(Y_test, Prediccion) #compare los datos
disp=metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
#curva de precisión del model 
from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_predictions (Y_test, Prediccion)
plt.show()
#valor 91% verdaderos positivos
```
